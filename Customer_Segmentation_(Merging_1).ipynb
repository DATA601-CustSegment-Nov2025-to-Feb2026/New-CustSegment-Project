## Join Transaction Dataset With Merged Dataset (Coupon & Coupon Description) & & Create a New Column 'received_campaign'

### Join Transaction Dataset With Merged Dataset (Coupon & Coupon Description)

```python
# Join 'campaign_df' with 'campaign_description_df' on campaign
transaction_campaign_df = transaction_df.merge(merged_campaign_df, on='household_key', how='left')

# Persist in memory
transaction_campaign_df = transaction_campaign_df.persist()
```

```python
# Display first 5 rows of the merged dataset
transaction_campaign_df.head(5)
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_transaction_campaign_df = transaction_campaign_df.shape[0].compute()
num_cols_transaction_campaign_df= len(transaction_campaign_df.columns)

print("Number of rows:", num_rows_transaction_campaign_df)
print("Number of columns:", num_cols_transaction_campaign_df)
```

### Create a New Column 'received_campaign' in Merged Dataset (Transaction, Coupon & Description)

```python
# Extract the start day from the 'day_range_tuple' column.
transaction_campaign_start_days = transaction_campaign_df['day_range_tuple'].map(
    lambda x: x[0] if isinstance(x, tuple) else np.nan,
    meta=('start_day', 'float64')
)

# Extract the end day from the 'day_range_tuple' column.
transaction_campaign_end_days = transaction_campaign_df['day_range_tuple'].map(
    lambda x: x[1] if isinstance(x, tuple) else np.nan,
    meta=('end_day', 'float64')
)

# Create a new column 'received_campaign' to indicate if 'day' in trasanction dataset falls within the campaign period (Yes or No)
transaction_campaign_df['received_campaign'] = (
    (transaction_campaign_df['day'] >= transaction_campaign_start_days) &
    (transaction_campaign_df['day'] <= transaction_campaign_end_days)
).map({True: 'yes', False: 'no'}, meta=('in_campaign', 'object'))

# Persist in memory
transaction_campaign_df = transaction_campaign_df.persist()

# Display first 5 rows after creating the new column
transaction_campaign_df.head()
```

```python
# Show rows where 'received_campaign' is 'yes'
transaction_campaign_yes_df = transaction_campaign_df [transaction_campaign_df['received_campaign'] == 'yes']

# View first 20 rows
transaction_campaign_yes_df.head()
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_transaction_campaign_yes_df = transaction_campaign_yes_df.shape[0].compute()
num_cols_transaction_campaign_yes_df = len(transaction_campaign_yes_df.columns)

print("Number of rows in 'transaction_campaign_yes_df' that show yes in 'received_campaign' column:", num_rows_transaction_campaign_yes_df)
print("Number of columns in 'transaction_campaign_yes_df' that show yes in 'received_campaign' column:", num_cols_transaction_campaign_yes_df)
```

### Join Transaction Dataset With 'transaction_campaign_yes' Dataset (Where 'received_campaign' is yes)

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_transaction_df = transaction_df.shape[0].compute()
num_cols_transaction_df = len(transaction_df.columns)

print("Number of rows in 'transaction_df':", num_rows_transaction_df)
print("Number of columns in 'transaction_df':", num_cols_transaction_df)
```

```python
# Join 'transaction_df' with 'transaction_campaign_df' on household key, product_id and day
transaction_transaction_campaign_yes_df = transaction_df.merge(
    transaction_campaign_yes_df,
    on=['household_key', 'product_id', 'day'],
    how='left'
)
```

```python
# Display first 5 rows of the merged dataset
transaction_transaction_campaign_yes_df.head()
```

```python
# Calculate the number of rows and columns in the Dask dataFrame after merging
num_rows_transaction_transaction_campaign_yes_df = transaction_transaction_campaign_yes_df.shape[0].compute()
num_cols_transaction_transaction_campaign_yes_df = len(transaction_transaction_campaign_yes_df.columns)

print("Number of rows:", num_rows_transaction_transaction_campaign_yes_df)
print("Number of columns:", num_cols_transaction_transaction_campaign_yes_df)
```

```python
# Select columns that do not end with "_y"
cols_to_keep_not_y = [col for col in transaction_transaction_campaign_yes_df.columns
                if not col.endswith('_y')]

# Keep only columns that do not end with "_y"
transaction_transaction_campaign_yes_df = transaction_transaction_campaign_yes_df[cols_to_keep_not_y]

# Drop the 'day_range_tuple' column
transaction_transaction_campaign_yes_df = transaction_transaction_campaign_yes_df.drop(columns='day_range_tuple')
```

```python
# Rename columns to remove the '_x' suffix added during merge
transaction_transaction_campaign_yes_df = (
    transaction_transaction_campaign_yes_df
    .rename(columns={col: col[:-2] for col in transaction_transaction_campaign_yes_df.columns
                     if col.endswith('_x')})
)
```

```python
transaction_transaction_campaign_yes_df.head()
```

```python
# Calculate the number of rows and columns in the Dask dataFrame after merging
num_rows_transaction_transaction_campaign_yes_df = transaction_transaction_campaign_yes_df.shape[0].compute()
num_cols_transaction_transaction_campaign_yes_df = len(transaction_transaction_campaign_yes_df.columns)

print("Number of rows:", num_rows_transaction_transaction_campaign_yes_df)
print("Number of columns:", num_cols_transaction_transaction_campaign_yes_df)
```

```python
# List all columns except 'campaign', 'description', 'start_day' and 'end_day'
cols_to_check = [col for col in transaction_transaction_campaign_yes_df.columns
                 if col not in ['campaign', 'description', 'start_day', 'end_day']]

# Create a boolean column marking duplicates within each partition
def mark_duplicates(df, subset):
    return df[df.duplicated(subset=subset, keep=False)]

# Apply to each partition
duplicates_transaction_transaction_campaign_yes_df = transaction_transaction_campaign_yes_df.map_partitions(
    mark_duplicates, subset=cols_to_check
)

# Compute number of duplicate rows
num_duplicates_transaction_transaction_campaign_yes_df = len(duplicates_transaction_transaction_campaign_yes_df.compute())

print(f"Number of duplicate rows: {num_duplicates_transaction_transaction_campaign_yes_df}")

duplicates_transaction_transaction_campaign_yes_df = duplicates_transaction_transaction_campaign_yes_df.persist()
```

```python
# Remove duplicates based on the subset of columns
transaction_campaign_no_duplicates_df = transaction_transaction_campaign_yes_df.drop_duplicates(
    subset=cols_to_check,
    keep='first'
)
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_transaction_campaign_no_duplicates_df = transaction_campaign_no_duplicates_df.shape[0].compute()
num_cols_transaction_campaign_no_duplicates_df = len(transaction_campaign_no_duplicates_df.columns)

print("Number of rows:", num_rows_transaction_campaign_no_duplicates_df)
print("Number of columns:", num_cols_transaction_campaign_no_duplicates_df)
```

```python
# Fill missing values in 'received_campaign' column with 'no'
transaction_campaign_no_duplicates_df['received_campaign'] = (
    transaction_campaign_no_duplicates_df['received_campaign']
    .fillna('no')
)
```

```python
# Count how many duplicate rows have display equal to "0" in 'duplicates_transaction_demographic_product_causal_df'
display_equal_0 = duplicates_transaction_demographic_product_causal_df[
    duplicates_transaction_demographic_product_causal_df['display'] == "0"
].shape[0].compute()

print("Number of duplicate rows have display equal to '0' in 'duplicates_transaction_demographic_product_causal_df':", display_equal_0, 'rows')
```

```python
# As our aim is to focus on transactional-level patterns for customer segmentation, the extra rows don't represent new purchases.
# They exist because a single transaction matched multiple display locations (the same product purchased in the same transaction may have different display locations.)
# Keeping these duplicates could inflate the number of transactions and bias our customer segmnetation, making it look like some customers buy more than they actually do.

# Remove all duplicate rows where display column equals to "0" (0 means not on display) in transaction_demographic_product_causal_df
# Convert Dask DataFrame to pandas
transaction_demographic_product_causal_df_pandas = transaction_demographic_product_causal_df.compute()

# Define columns to check for duplicates (excluding 'display' and 'mailer')
cols_to_identify_transaction_demographic_product_causal_df_pandas = [col for col in transaction_demographic_product_causal_df_pandas.columns if col not in ['display', 'mailer']]

# Remove duplicate rows where display column equals to "0" (0 means not on display) in transaction_demographic_product_causal_df
transaction_demographic_product_causal_df_pandas = transaction_demographic_product_causal_df_pandas[~(
    (transaction_demographic_product_causal_df_pandas.duplicated(subset=cols_to_identify_transaction_demographic_product_causal_df_pandas, keep=False)) &
    (transaction_demographic_product_causal_df_pandas['display'] == "0")
)]

# Convert back to Dask DataFrame
transaction_demographic_product_causal_df = dd.from_pandas(transaction_demographic_product_causal_df_pandas)

# Persist in memory
transaction_demographic_product_causal_df = transaction_demographic_product_causal_df.persist()
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_transaction_demographic_product_causal_df = transaction_demographic_product_causal_df.shape[0].compute()
num_cols_transaction_demographic_product_causal_df = len(transaction_demographic_product_causal_df.columns)

print("Number of rows in transaction_demographic_product_causal_df:", num_rows_transaction_demographic_product_causal_df)
print("Number of columns in transaction_demographic_product_causal_df:", num_cols_transaction_demographic_product_causal_df)
```
