# Joining Dataset

## Join Transaction Dataset With Demographic Dataset

```python
# Join 'transaction_df' with 'demographic_df' on household_key
transaction_demographic_df = transaction_df.merge(demographic_df, on='household_key', how='left')

# Show all columns
pd.set_option('display.max_columns', None)

# Display first 5 rows of the merged dataset
transaction_demographic_df.head()
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_transaction_demographic_df = transaction_demographic_df.shape[0].compute()
num_cols_transaction_demographic_df = len(transaction_demographic_df.columns)

print("Number of rows:", num_rows_transaction_demographic_df)
print("Number of columns:", num_cols_transaction_demographic_df)
```

## Join Merged Dataset (Transaction & Demographic) With Product Dataset

```python
# Join 'transaction_demographic_df' with 'product_df' on product_id
transaction_demographic_product_df = transaction_demographic_df.merge(product_df, on='product_id', how='left')

# Show all columns
pd.set_option('display.max_columns', None)

# Display first 5 rows of the merged dataset
transaction_demographic_product_df.head()
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_transaction_demographic_product_df = transaction_demographic_product_df.shape[0].compute()
num_cols_transaction_demographic_product_df = len(transaction_demographic_product_df.columns)

print("Number of rows:", num_rows_transaction_demographic_product_df)
print("Number of columns:", num_cols_transaction_demographic_product_df)
```

## Join Merged Dataset (Transaction, Demographic & Product) With Causal Dataset

```python
# Join 'transaction_demographic_product_df' with 'causal_df' on product_id, store_id and week_no
transaction_demographic_product_causal_df = transaction_demographic_product_df.merge(causal_df, on=['product_id', 'store_id', 'week_no'], how='left')

# Persist in memory
transaction_demographic_product_causal_df = transaction_demographic_product_causal_df.persist()
```

```python
# Show all columns
pd.set_option('display.max_columns', None)

# Display first 5 rows of the merged dataset
transaction_demographic_product_causal_df.head()
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_transaction_demographic_product_causal_df = transaction_demographic_product_causal_df.shape[0].compute()
num_cols_transaction_demographic_product_causal_df = len(transaction_demographic_product_causal_df.columns)

print("Number of rows in transaction_demographic_product_causal_df:", num_rows_transaction_demographic_product_causal_df)
print("Number of columns in transaction_demographic_product_causal_df:", num_cols_transaction_demographic_product_causal_df)
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_transaction_demographic_product_df = transaction_demographic_product_df.shape[0].compute()
num_cols_transaction_demographic_product_df = len(transaction_demographic_product_df.columns)

print("Number of rows in transaction_demographic_product_df:", num_rows_transaction_demographic_product_df)
print("Number of columns in transaction_demographic_product_df:", num_cols_transaction_demographic_product_df)
```

```python
# Calculate the number of row difference between 'transaction_demographic_product_causal_df' and 'transaction_demographic_product_df'
row_difference_1 = num_rows_transaction_demographic_product_causal_df - num_rows_transaction_demographic_product_df

print(f"Row difference between 'transaction_demographic_product_causal_df' and 'transaction_demographic_product_df': {row_difference_1} rows")
```

```python
# List all columns except 'display' and 'mailer'
cols_to_check = [col for col in transaction_demographic_product_causal_df.columns
                 if col not in ['display', 'mailer']]

# Create a boolean column marking duplicates within each partition
def mark_duplicates(df, subset):
    return df[df.duplicated(subset=subset, keep=False)]

# Apply to each partition
duplicates_transaction_demographic_product_causal_df = transaction_demographic_product_causal_df.map_partitions(
    mark_duplicates, subset=cols_to_check
)

# Compute number of duplicate rows
num_duplicates_transaction_demographic_product_causal_df  = len(duplicates_transaction_demographic_product_causal_df.compute())

print(f"Number of duplicate rows: {num_duplicates_transaction_demographic_product_causal_df}")
```

```python
# Display first 5 rows of the merged dataset
duplicates_transaction_demographic_product_causal_df.head(5)

# After the left join, the the number of rows increased from 2,595,732 to 2,596,590 (858 rows).
# This is due to multiple matching rows for 'display' and 'mailer' columns in the right table (causal_df).
```

```python
# Count how many duplicate rows have display equal to "0" in 'duplicates_transaction_demographic_product_causal_df'
display_equal_0 = duplicates_transaction_demographic_product_causal_df[
    duplicates_transaction_demographic_product_causal_df['display'] == "0"
].shape[0].compute()

print("Number of duplicate rows have display equal to '0' in 'duplicates_transaction_demographic_product_causal_df':", display_equal_0, 'rows')
```

```python
# As our aim is to focus on transactional-level patterns for customer segmentation, the extra rows don't represent new purchases.
# They exist because a single transaction matched multiple display locations (the same product purchased in the same transaction may have different display locations.)
# Keeping these duplicates could inflate the number of transactions and bias our customer segmnetation, making it look like some customers buy more than they actually do.

# Remove all duplicate rows where display column equals to "0" (0 means not on display) in transaction_demographic_product_causal_df
# Convert Dask DataFrame to pandas
transaction_demographic_product_causal_df_pandas = transaction_demographic_product_causal_df.compute()

# Define columns to check for duplicates (excluding 'display' and 'mailer')
cols_to_identify_transaction_demographic_product_causal_df_pandas = [col for col in transaction_demographic_product_causal_df_pandas.columns if col not in ['display', 'mailer']]

# Remove duplicate rows where display column equals to "0" (0 means not on display) in transaction_demographic_product_causal_df
transaction_demographic_product_causal_df_pandas = transaction_demographic_product_causal_df_pandas[~(
    (transaction_demographic_product_causal_df_pandas.duplicated(subset=cols_to_identify_transaction_demographic_product_causal_df_pandas, keep=False)) &
    (transaction_demographic_product_causal_df_pandas['display'] == "0")
)]

# Convert back to Dask DataFrame
transaction_demographic_product_causal_df = dd.from_pandas(transaction_demographic_product_causal_df_pandas)

# Persist in memory
transaction_demographic_product_causal_df = transaction_demographic_product_causal_df.persist()
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_transaction_demographic_product_causal_df = transaction_demographic_product_causal_df.shape[0].compute()
num_cols_transaction_demographic_product_causal_df = len(transaction_demographic_product_causal_df.columns)

print("Number of rows in transaction_demographic_product_causal_df:", num_rows_transaction_demographic_product_causal_df)
print("Number of columns in transaction_demographic_product_causal_df:", num_cols_transaction_demographic_product_causal_df)
```

## Join Merged Dataset (Transaction, Demographic, Product & Causal) With Merged Coupon Dataset

```python
# Join 'transaction_demographic_product_causal_df' with 'merged_coupon_df' on household_key, day and product_id.
merged_transaction_coupon_df = transaction_demographic_product_causal_df.merge(merged_coupon_df, on=['household_key', 'day', 'product_id'], how='left')

# Persist in memory
merged_transaction_coupon_df = merged_transaction_coupon_df.persist()
```

```python
# Show all columns
pd.set_option('display.max_columns', None)

# Display first 5 rows of the merged dataset
merged_transaction_coupon_df.head(5)
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_merged_transaction_coupon_df = merged_transaction_coupon_df.shape[0].compute()
num_cols_merged_transaction_coupon_df = len(merged_transaction_coupon_df.columns)

print("Number of rows in merged_transaction_coupon_df:", num_rows_merged_transaction_coupon_df)
print("Number of columns in merged_transaction_coupon_df:", num_cols_merged_transaction_coupon_df)
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_transaction_demographic_product_causal_df = transaction_demographic_product_causal_df.shape[0].compute()
num_cols_transaction_demographic_product_causal_df = len(transaction_demographic_product_causal_df.columns)

print("Number of rows in transaction_demographic_product_causal_df:", num_rows_transaction_demographic_product_causal_df)
print("Number of columns in transaction_demographic_product_causal_df:", num_cols_transaction_demographic_product_causal_df)
```

```python
# Calculate the number of row difference between 'transaction_demographic_product_causal_df' and 'merged_transaction_coupon_df'
row_difference_2 = num_rows_merged_transaction_coupon_df - num_rows_transaction_demographic_product_causal_df

print(f"Row difference between 'transaction_demographic_product_causal_df' and 'merged_transaction_coupon_df': {row_difference_2} rows")
```

```python
# List all columns except 'display' and 'mailer'
cols_to_check = [col for col in merged_transaction_coupon_df.columns
                 if col not in ['coupon_upc', 'campaign', 'household_key']]

# Create a boolean column marking duplicates within each partition
def mark_duplicates(df, subset):
    return df[df.duplicated(subset=subset, keep=False)]

# Apply to each partition
duplicates_merged_transaction_coupon_df = merged_transaction_coupon_df.map_partitions(
    mark_duplicates, subset=cols_to_check
)

# Compute number of duplicate rows
num_duplicates_merged_transaction_coupon_df = len(duplicates_merged_transaction_coupon_df.compute())

print(f"Number of duplicate rows: {num_duplicates_merged_transaction_coupon_df}")
```

```python
# Convert 'coupon_upc' column from gloat to string to avoid scientific notation caused by large numeric ID.
duplicates_merged_transaction_coupon_df['coupon_upc'] = duplicates_merged_transaction_coupon_df['coupon_upc'].astype('str')

# Display first 5 rows of the merged dataset
duplicates_merged_transaction_coupon_df.head(5)

# After the left join, the the numner of rows increased from 2,596,590 to 2,596,722 (182 rows).
# This is due to multiple matching rows for 'coupon_upc' and 'campaign' columns in the right table (merged_transaction_coupon_df).
```

```python
# As our aim is to focus on transactional-level patterns for customer segmentation, the extra rows don't represent new purchases.
# They exist because a single transaction matched multiple coupons or campaigns (a customer may have used or been eligible for multiple coupons or campaigns on the same product purchase.)
# Keeping these duplicates could inflate the number of transactions and bias our customer segmnetation, making it look like some customers buy more than they actually do.

# Define columns to check for duplicates (excluding 'coupon_upc' and 'campaign')
cols_to_identify_merged_transaction_coupon_df = [
    'household_key', 'basket_id', 'day', 'product_id', 'quantity',
    'sales_value', 'store_id', 'retail_disc', 'trans_time', 'week_no', 'coupon_disc',
    'coupon_match_disc', 'age_desc', 'marital_status_code', 'income_desc', 'homeowner_desc',
    'hh_comp_desc', 'household_size_desc', 'kid_category_desc',
    'manufacturer', 'department', 'brand', 'commodity_desc',
    'sub_commodity_desc', 'curr_size_of_product', 'display', 'mailer'
]

# We want to keep the coupon with the highest dicount
# Sort by discount descending and keep first per transaction
merged_transaction_coupon_df = merged_transaction_coupon_df.sort_values('coupon_disc', ascending=True)
merged_transaction_coupon_df_no_dup = merged_transaction_coupon_df.drop_duplicates(subset=cols_to_identify_merged_transaction_coupon_df , keep='first')

# Persist in memory
merged_transaction_coupon_df_no_dup = merged_transaction_coupon_df_no_dup.persist()

# Display first 5 rows of the merged dataset
merged_transaction_coupon_df_no_dup.head(5)
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_merged_transaction_coupon_df_no_dup = merged_transaction_coupon_df_no_dup.shape[0].compute()
num_cols_merged_transaction_coupon_df_no_dup = len(merged_transaction_coupon_df_no_dup.columns)

print("Number of rows in merged_transaction_coupon_df:", num_rows_merged_transaction_coupon_df_no_dup)
print("Number of columns in merged_transaction_coupon_df:", num_cols_merged_transaction_coupon_df_no_dup)
```

## Join Merged Transaction & Coupon Dataset With Campaign Description Dataset

```python
# Join 'merged_transaction_coupon_df_no_dup' with 'campaign_description_df' on campaign
merged_transaction_coupon_campaign_df = merged_transaction_coupon_df_no_dup.merge(campaign_description_df, on=['campaign'], how='left')
final_merged_df = merged_transaction_coupon_campaign_df

# Persist in memory
final_merged_df = final_merged_df.persist()
```

```python
# Calculate the number of rows and columns in the Dask dataFrame
num_rows_final_merged_df = final_merged_df.shape[0].compute()
num_cols_final_merged_df = len(final_merged_df.columns)

print("Number of rows in merged_transaction_coupon_df:", num_rows_final_merged_df)
print("Number of columns in merged_transaction_coupon_df:", num_cols_final_merged_df)
```
